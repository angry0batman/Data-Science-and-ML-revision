```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[3.4,5.5,6.7]
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[30,55,22,43]
color=["red","blue","green","black"]
plt.scatter(x,y,c=color)
plt.title("Scatter Plot")
plt.xlabel("Days")
plt.ylabel("No of customers")
plt.show()
```

```python
x=[1,2,3,4]
y=[30,55,22,43]
color=["red","blue","green","black"]
plt.scatter(x,y,c=color,s=400)
plt.title("Scatter Plot")
plt.xlabel("Days")
plt.ylabel("No of customers")
plt.show()
```

```python
x=[1,2,3,4]
y=[30,55,22,43]
color=["red","blue","green","black"]
size=[400,100,250,300]
plt.scatter(x,y,c=color,s=size)
plt.title("Scatter Plot")
plt.xlabel("Days")
plt.ylabel("No of customers")
plt.show()
```

```python
x=[1,2,3,4]
y=[30,55,22,43]
color=["red","blue","green","black"]
size=[400,100,250,300]
plt.scatter(x,y,c=color,s=size,alpha=0.6)
plt.title("Scatter Plot")
plt.xlabel("Days")
plt.ylabel("No of customers")
plt.show()
```

```python
x=[1,2,3,4]
y=[30,55,22,43]
color=["red","blue","green","black"]
size=[400,100,250,300]
plt.scatter(x,y,c=color,s=size,marker="*")
plt.title("Scatter Plot")
plt.xlabel("Days")
plt.ylabel("No of customers")
plt.show()
```
