```python
x=["reliance","adani","ongc","hindustan petroleums"]
y=[100,200,75,400]
z=[50,150,90,300]
c=['red','yellow','blue','green']
d=['yellow','red','black','blue']
plt.title("Company v/s Growth Rate")
plt.xlabel("Companies")
plt.ylabel('Growth percentage this session')
plt.bar(x,y,color=c,width=0.4,edgecolor="black",label='Growth Rate1')
plt.bar(x,z,color=d,width=0.4,edgecolor="black",label='Growth Rate2')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = ["reliance", "adani", "ongc", "hindustan petroleums"]
y = [100, 200, 75, 400]
z = [50, 150, 90, 300]
c = ['red', 'yellow', 'blue', 'green']
d = ['yellow', 'red', 'black', 'blue']

# Set the width of the bars
width = 0.4

# Create an array with the positions of the bars on the x-axis
x_indices = np.arange(len(x))

# Plot the bars
plt.bar(x_indices - width/2, y, width=width, color=c, edgecolor="black", label='Growth Rate1')
plt.bar(x_indices + width/2, z, width=width, color=d, edgecolor="black", label='Growth Rate2')

# Customize the axes and labels
plt.title("Company v/s Growth Rate")
plt.xlabel("Companies")
plt.ylabel("Growth percentage this session")
plt.xticks(x_indices, x)
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = ["reliance", "adani", "ongc", "hindustan petroleums"]
y = [100, 200, 75, 400]
z = [50, 150, 90, 300]
c = ['red', 'yellow', 'blue', 'green']
d = ['yellow', 'red', 'black', 'blue']



# Plot the bars
plt.barh(x , y, color=c, edgecolor="black", label='Growth Rate1')
plt.barh(x, z, color=d, edgecolor="black", label='Growth Rate2')

# Customize the axes and labels
plt.title("Company v/s Growth Rate")
plt.xlabel("Companies")
plt.ylabel("Growth percentage this session")

plt.legend()

# Show the plot
plt.show()
```
