Bar Plot

```python
import matplotlib.pyplot as plt

x=["reliance","adani","ongc","hindustan petroleums"]
y=[100,200,75,400]
c=['red','yellow','blue','green']
plt.title("Company v/s Growth Rate")
plt.xlabel("Companies")
plt.ylabel('Growth percentage this session')
plt.bar(x,y,color=c)
plt.show()
```
